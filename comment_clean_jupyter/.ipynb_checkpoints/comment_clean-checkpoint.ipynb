```python
import pymysql
import pandas as pd
import re
from bs4 import BeautifulSoup
from hanziconv import HanziConv
from zhon import hanzi
import string
punctuation=hanzi.punctuation + string.punctuation
punctuation = set([i for i in punctuation])
import html
import pymysql
```

```python
conn = pymysql.connect(host='127.0.0.1',port=3306,user='root',password='eXYhzAWjyvy8grwM',db='game_process',charset='utf8mb4')
sql = "SELECT id,source,game_name, user_name,comment_time, content, score FROM s_game_comments_qq WHERE source='qq_mobile'"
df = pd.read_sql(sql, conn)  # 返回的是：
conn.close()
```

```python
def extract_text(x):
    soup = BeautifulSoup(x, 'html.parser')
    return soup.text

def clear_text(x):
    x = html.unescape(x)  # 反转义字符，显示真实内容。
    x = HanziConv.toSimplified(x) # 汉字繁体 转成 简体
    x = x.strip() # 去除字符 首尾的指定 字符
    return x

def remove_punctuation(x):  # 删除标点
    x = "".join([i for i in x if i not in punctuation]) # 遍历评论每个字符，若是标点，去掉，用空字符连接
    return re.sub(r'\s+', ' ', x) # 把 x 通过正则表达式 (\s 代表字母 + 代表多个)，替换成 空字符
```

```python
df['game_name'] = df['game_name'].apply(clear_text)
df['content'] = df['content'].apply(clear_text)
```

```python
df.head()
```

```python
df['comment_time']=df['comment_time'].astype('int32')
df['score']=df['score'].astype('float64')
```

```python
df['content_len'] = df['content'].apply(len)
```

```python
def get_help_content(x):
    x = "".join([i for i in x if i not in punctuation])
    x = re.sub(r'\s+', '', x)
    x = x[:int(len(x) * 0.7)]
    return x
```

```python
df['help_content'] = df['content'].apply(get_help_content)
```

```python
numbers = [  r'1234567890',
            r'一二三四五六七八九',
            r'一二三四五六七八九',
            r'①②③④⑤⑥⑦⑧⑨⑩⑪⑫⑬⑭⑮⑯⑰⑱⑲⑳',
            r'⑴⑵⑶⑷⑸⑹⑺⑻⑼⑽⑾⑿⒀⒁⒂⒃⒄⒅⒆⒇',
            r'⒈⒉⒊⒌⒋⒌⒍⒎⒏⒐⒑⒒⒓⒔⒕⒖⒗⒘⒙⒚⒛',
            r'❶❷❸❹❺❻❼❽❾❿',
            r'１２３４５６７８９',
            r'㈠㈡㈢㈣㈤㈥㈦㈧㈧㈨㈩',
            r'ＯＩ',
         ]
numbers = set([n2 for n1 in numbers for n2 in n1])
def number_count(x):
    global numbers
    return len([i  for i in x if i in numbers])
```

```python
df['ncount'] = df['help_content'].apply(number_count)
```

```python
df[df['ncount'] == True].head()
```

```python
cc = df[['help_content']].groupby(['help_content']).apply(len).reset_index(level = ['help_content'])
cc.columns = ['help_content', 'ccount']
cc.head()
```

```python
uc = df[['user_name']].groupby(['user_name']).apply(len).reset_index(level = ['user_name'])
uc.columns = ['user_name', 'ucount']
uc.head()
```

```python
df = df.merge(cc, on = 'help_content').merge(uc, on = 'user_name')
```
